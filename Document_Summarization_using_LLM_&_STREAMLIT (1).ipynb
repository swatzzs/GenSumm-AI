```python
%%writefile app.py
import streamlit as st
from transformers import pipeline, MarianMTModel, MarianTokenizer
from gtts import gTTS
import pdfplumber
import asyncio
import time
from concurrent.futures import ThreadPoolExecutor

# Load models once and cache them
@st.cache_resource
def load_models():
    summarizer = pipeline("summarization", model="facebook/bart-base")  # Use a smaller model for faster performance
    qa_pipeline = pipeline("question-answering", model="distilbert/distilbert-base-cased-distilled-squad")
    return summarizer, qa_pipeline

summarizer, qa_pipeline = load_models()

# Supported languages and corresponding MarianMT model codes
LANGUAGE_MODELS = {
    'Spanish': 'Helsinki-NLP/opus-mt-en-es',
    'French': 'Helsinki-NLP/opus-mt-en-fr',
    'German': 'Helsinki-NLP/opus-mt-en-de',
    'Italian': 'Helsinki-NLP/opus-mt-en-it',
    'Portuguese': 'Helsinki-NLP/opus-mt-en-pt',
    'Hindi': 'Helsinki-NLP/opus-mt-en-hi'  # Added Hindi
}

# Function to extract text from PDF
@st.cache_data
def extract_text_from_pdf(pdf_file):
    text = ""
    with pdfplumber.open(pdf_file) as pdf:
        for page in pdf.pages:
            text += page.extract_text() or ""
    return text

# Function to summarize a chunk of text
def summarize_chunk(chunk, max_summary_length):
    try:
        summary = summarizer(chunk, max_length=max_summary_length, min_length=20, do_sample=False)[0]['summary_text']
        return summary
    except Exception as e:
        st.error(f"Error summarizing chunk: {e}")
        return ""

# Asynchronous summarization function with limited concurrency
async def summarize_text(text, max_chunk_size=1024, max_summary_length=50):
    chunks = [text[i:i + max_chunk_size] for i in range(0, len(text), max_chunk_size)]
    with ThreadPoolExecutor(max_workers=4) as executor:
        loop = asyncio.get_event_loop()
        summaries = await asyncio.gather(
            *[loop.run_in_executor(executor, summarize_chunk, chunk, max_summary_length) for chunk in chunks]
        )
    return " ".join(summaries)

# Asynchronous translation function with batch processing
async def translate_text(text, model_name, chunk_size=1000):
    translation_model = MarianMTModel.from_pretrained(model_name)
    tokenizer = MarianTokenizer.from_pretrained(model_name)

    translated_text_full = ""
    chunks = [text[i:i + chunk_size] for i in range(0, len(text), chunk_size)]
    for text_chunk in chunks:
        try:
            loop = asyncio.get_event_loop()
            translated_tokens = await loop.run_in_executor(None, lambda: translation_model.generate(**tokenizer(text_chunk, return_tensors="pt", padding=True, clean_up_tokenization_spaces=True)))
            translated_text_full += tokenizer.decode(translated_tokens[0], skip_special_tokens=True) + "\n"
        except Exception as e:
            st.error(f"Error translating chunk: {e}")
    return translated_text_full

# Function to convert text to audio
def convert_text_to_audio(text, audio_path, lang='es'):
    try:
        tts = gTTS(text, lang=lang)
        tts.save(audio_path)
    except Exception as e:
        st.error(f"Error converting text to audio: {e}")

# Wrapper for asynchronous summarization
def get_summary(text, max_summary_length):
    return asyncio.run(summarize_text(text, max_summary_length=max_summary_length))

# Wrapper for asynchronous translation
def get_translation(text, model_name):
    return asyncio.run(translate_text(text, model_name))

# Streamlit UI
st.title('PDF Processing and Query Bot Application')

uploaded_file = st.file_uploader("Choose a PDF file", type="pdf")

if uploaded_file is not None:
    with st.spinner("Extracting text from PDF..."):
        start_time = time.time()
        text = extract_text_from_pdf(uploaded_file)
        st.write("**Extracted Text:**")
        st.text_area("Extracted Text", text, height=300)
        st.write(f"Extraction Time: {time.time() - start_time:.2f} seconds")

        if not text.strip():
            st.error("The extracted text is empty or unreadable. Please check the PDF file.")
        else:
            st.write("**Summary Settings**")
            max_chunk_size = st.slider("Select chunk size for summarization (in characters):", 512, 2048, 1024)
            max_summary_length = st.slider("Select maximum summary length:", 20, 100, 50)

            st.write("**Summary:**")
            with st.spinner("Summarizing text..."):
                start_time = time.time()
                summary = get_summary(text, max_summary_length=max_summary_length)
                st.write(summary)
                st.write(f"Summarization Time: {time.time() - start_time:.2f} seconds")

            # Export Summary Option
            st.download_button(label="Download Summary", data=summary, file_name="summary.txt")

            st.write("**Translation Settings**")
            selected_language = st.selectbox("Choose a target language for translation:", list(LANGUAGE_MODELS.keys()))

            st.write("**Translation (to selected language):**")
            with st.spinner(f"Translating text to {selected_language}..."):
                start_time = time.time()
                translated_text = get_translation(text, LANGUAGE_MODELS[selected_language])
                st.text_area(f"Translated Text ({selected_language})", translated_text, height=300)
                st.write(f"Translation Time: {time.time() - start_time:.2f} seconds")

            # Export Translation Option
            st.download_button(label="Download Translated Text", data=translated_text, file_name=f"translated_text_{selected_language}.txt")

            # Convert Translated Text to Audio
            audio_path = "/tmp/translated_output.mp3"
            lang_code = 'hi' if selected_language == 'Hindi' else 'es'  # Language code for text-to-speech
            convert_text_to_audio(translated_text, audio_path, lang=lang_code)
            st.audio(audio_path, format="audio/mp3")

            # Export Audio Option
            with open(audio_path, "rb") as audio_file:
                st.download_button(label="Download Audio", data=audio_file, file_name="translated_audio.mp3", mime="audio/mp3")

            st.subheader('Ask a question about the document')
            user_query = st.text_input("Your question:", "")

            if st.button('Submit'):
                if user_query:
                    start_time = time.time()
                    try:
                        if text.strip() == "":
                            st.write("The extracted text is empty. Please check the PDF file.")
                        else:
                            result = qa_pipeline(question=user_query, context=text)
                            answer = result.get('answer', 'No answer found.')
                            elapsed_time = time.time() - start_time
                            st.write(f"**Answer:** {answer}")
                            st.write(f"**Query Time:** {elapsed_time:.2f} seconds")
                    except Exception as e:
                        st.error(f"Error getting answer: {e}")
                else:
                    st.write("Please enter a question.")
```
